```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y1=[i*2 for i in x]
y2=[i**2 for i in x]
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('double')
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("square")
plt.show()
```

```python
z=[1,2,3,4,5,6]
a1=[i*2 for i in x]
a2=[i**2 for i in x]
a3=[i*6 for i in x]
a4=[i**0.5 for i in x]
```

```python
plt.subplot(1,3,1)
plt.plot(x,a1)
plt.subplot(1,3,2)
plt.plot(x,a2)
plt.subplot(1,3,3)
plt.plot(x,a3)
```

```python
plt.subplot(2,2,1)
plt.title("double the number")
plt.plot(z,a1)
plt.grid(True)
plt.subplot(2,2,2)
plt.title("square")
plt.plot(z,a2)
plt.grid(True)
plt.subplot(2,2,3)
plt.title("mul of 6")
plt.grid(True)
plt.plot(z,a3)
plt.subplot(2,2,4)
plt.title("square root")
plt.plot(z,a4)
plt.tight_layout()
plt.grid(True)
plt.show()

```

```python
fig,axs=plt.subplots(1,2,figsize=(10,4))
axs[0].plot(z,a1)
axs[0].set_title('double')
axs[1].plot(z,a2)
axs[1].set_title("square")
fig.suptitle("my comparison plot",fontsize=20)
plt.show()
```
