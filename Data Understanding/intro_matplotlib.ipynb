```python
years=[1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs=[500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.xlabel("years")
plt.ylabel("Runs Score")
plt.plot(years,runs)
plt.title("Sachin Tendulkar's Yearly Score")
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years,kohli,label="virat kohli")
plt.plot(years,sehwag,label="virender sehwag")
plt.xlabel("year")
plt.ylabel("Runs scored")
plt.title("Performance Comparision")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli,'ro--',label="kohli")
plt.plot(years,sehwag,'g^:',label="sehwag")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli,'y1--',label="kohli")
plt.plot(years,sehwag,'bD:',label="sehwag")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli,'m^-',linewidth="1",label="Virat kohli")
plt.plot(years,sehwag,'kD:',linewidth="2",label="virander sehwag")
plt.grid(True)
plt.legend()
plt.show()
```
