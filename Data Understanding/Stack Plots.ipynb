```python
import matplotlib.pyplot as plt
```

```python
activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5] 
```

```python
color=['blue','red','yellow','pink']
plt.figure(figsize=(6,6))
plt.pie(time_spent,labels=activities,colors=color,startangle=90,autopct='%1.1f%%',shadow=True)
plt.title("school going student")
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
```

```python
labels=['studying','playing','watching_tv','sleeping']
colors=['#C9C312','#0FBF7C','#0F90BF','#B860F0']
plt.figure(figsize=(6,6))
plt.stackplot(days,studying,playing,watching_tv,sleeping,labels=labels,colors=colors,alpha=0.4)
plt.legend(loc="upper left")
plt.title("After School Activities")
plt.xlabel('Days')
plt.ylabel('Hours')
plt.grid(True)
```
