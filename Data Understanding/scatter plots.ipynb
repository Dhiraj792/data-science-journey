```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
```

```python
colors=['red' if score<60 else'green'for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=colors)
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
colors=['red' if score<60 else'green'for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='viridis')
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
colors=['red' if score<60 else'green'for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='viridis')
plt.colorbar(label='score')
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
colors=['red' if score<60 else'green'for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='binary')
plt.colorbar(label='score')
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
colors=['red' if score<60 else'green'for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='hot')
plt.colorbar(label='score')
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
colors=['red' if score<60 else'green'for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='autumn')
plt.colorbar(label='score')
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
colors=['red' if score<60 else'green'for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,cmap='plasma')
plt.colorbar(label='score')
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
scores=[30,67,89,134,654,844,453,301,688]
plt.scatter(study_hours,exam_scores,c=scores,cmap='plasma')
plt.colorbar(label='exam_scores')
plt.title("marks according to study")
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
scores=[30,67,89,134,654,844,453,301,688]
plt.scatter(study_hours,exam_scores,c=scores,cmap='plasma')
plt.colorbar(label='exam_scores')
plt.title("marks according to study")
for i in range(len(study_hours)):
    plt.annotate(f's{i+1}',(study_hours[i]+0.1,exam_scores[i]-1.5))
plt.grid(True)
plt.xlabel("study hours")
plt.ylabel("Marks")
plt.show()
```

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
```

```python
plt.scatter(class_a_hours,class_a_scores,label='A',color='blue')
plt.scatter(class_b_hours,class_b_scores,label='B',color='red')
plt.title("Scores Of Two Different Classes")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
