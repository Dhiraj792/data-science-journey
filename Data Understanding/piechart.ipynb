```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
# Data
data = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
```

```python
plt.title("career runs of indian batsmen")
colors=['#C9C312','#0FBF7C','#0F90BF','#B860F0']
plt.pie(runs,labels=data,colors=colors)
plt.tight_layout()
plt.show()
```

```python
plt.pie(runs,labels=data,colors=colors,
        wedgeprops={'edgecolor':'#7F10E6','linewidth':2,'linestyle':'--'})
plt.show()
```

```python
explode=[0.1,0.05,0,0]
plt.title("career runs of indian batsmen")
colors=['#C9C312','#0FBF7C','#0F90BF','#B860F0']
plt.pie(runs,labels=data,explode=explode,colors=colors)
plt.tight_layout()
plt.show()
```

```python
explode=[0.1,0.05,0,0]
plt.title("career runs of indian batsmen")
colors=['#C9C312','#0FBF7C','#0F90BF','#B860F0']
plt.pie(runs,labels=data,explode=explode,colors=colors,)
plt.tight_layout()
plt.show()
```

```python
explode=[0.1,0.05,0,0]
plt.title("career runs of indian batsmen")
colors=['#C9C312','#0FBF7C','#0F90BF','#B860F0']
plt.pie(runs,labels=data,explode=explode,colors=colors,shadow=True,startangle=130)
plt.tight_layout()
plt.show()
```

```python
explode=[0.1,0.05,0,0]
plt.title("career runs of indian batsmen")
colors=['#C9C312','#0FBF7C','#0F90BF','#B860F0']
plt.pie(runs,labels=data,explode=explode,colors=colors,shadow=True,autopct='%1.1f%%')
plt.tight_layout()
plt.show()
```
